# 2 Data preprocessing - stage 2

<b> Purpose of the action </b> - transform some numerical features with a weak or random correlation to the target value into categorical features

<b> </b>
<b> Action plan </b>:
- Check the correlation between the original features and the target value using Pearson's correlation and graphical tools
- Select weak or random correlated features
- Transform numeric features to categorical features using different numbers of bins (categories, intervals)
- Select optimal numbers of bins using Pearson's correlation and graphical interpretation
- Add new features to datasets

## 2.1 Data loading

### 2.1.1 Import nessesary libraries

```python
import numpy as np
import pandas as pd
from data_preprocessing import calculate_and_show_correlation_bins, numeric_to_categorical
```

### 2.1.2 Read the data

```python
train_set = pd.read_csv("./preprocessed_data/train_set_stage1.csv", index_col=0)
train_set.head()
```

```python
test_set = pd.read_csv("./preprocessed_data/test_set_stage1.csv", index_col=0)
test_set.reset_index(inplace=True, drop=True)
test_set.head()
```

## 2.2 Check the correlation between the original features and the target mean value 

### 2.2.1 Define list of column to chceck 

```python
# look on all column
print(list(train_set.columns))
```

```python
# skip target, helpers and binary columns
features_to_check = [
                        'HomeTeamWinRatio', 'AwayTeamWinRatio',
                        'HomeTeamGoalsScored', 'AwayTeamGoalsScored', 'HomeTeamGoalsLost',
                        'AwayTeamGoalsLost', 'HomeTeamShootsMade', 'AwayTeamShootsMade',
                        'HomeTeamTargetShootsMade', 'AwayTeamTargetShootsMade',
                        'HomeTeamCorners', 'AwayTeamCorners', 'HomeTeamTotalPoints',
                        'AwayTeamTotalPoints', 'HomeTeamPointsFromLast3Matches',
                        'AwayTeamPointsFromLast3Matches', 'HomeTeamPointsFromLast5Matches',
                        'AwayTeamPointsFromLast5Matches', 'HomeTeamPointsFromLast10Matches',
                        'AwayTeamPointsFromLast10Matches', 'HomeTeamLastYearPosition', 
                        'AwayTeamLastYearPosition','HomeTeamGoalsDifference', 
                        'AwayTeamGoalsDifference','TotalGoalsDifference',
                        'DifferenceTotalPoints','Difference1MatchPoints', 
                        'Difference3MatchesPoints','Difference5MatchesPoints',
                        'Difference10MatchesPoints','DifferenceInShoots', 
                        'DifferenceInTargetShoots', 'DifferenceInCorners',
                        'DifferenceInLastYearPosition'
                    ]
```

### 2.2.2 Calculate and show correlation between features and target mean value

```python
# store Pearson's coeffs for all features in numpy array
corr_coefs = np.array([calculate_and_show_correlation(train_set, 'FTR', feature) for feature in features_to_check])
```

## 2.3 Select weak or random correlated features

Use columns with a correlation coefficient less than 0.9 for transformation

```python
mask = np.abs(corr_coefs) < 0.9
features_to_transform = np.array(features_to_check)[mask]
features_to_transform
```

## 2.4 Select optimal number of bins for numeric features (multivalued)

Plot correlation between all features from above list and mean target value

```python
# define list of numbers of bins to check
bins_set = [6, 8, 10, 12, 14]

# store results in list
result_sets = [calculate_and_show_correlation_bins(train_set, 'FTR', bins_set, feature) for feature in features_to_transform]
```

```python
# calculate optimal number of bins for all features based on Pearson's coefficient
optimal_bins = [result_sets[i][:,1:][np.abs(result_sets[i][:,1]).argmax()][1] for i in range(len(features_to_transform))]
optimal_bins
```

```python
features_to_transform
```

```python
# change some values manually based on plots
optimal_bins[3] = 8
optimal_bins[4] = 8
optimal_bins[5] = 8
optimal_bins[6] = 6
optimal_bins[9] = 6
optimal_bins[15] = 8
optimal_bins[16]= 10
optimal_bins[18]= 10

optimal_bins
```

## 2.5 Transform selected numeric features to categorical features with appropriate numbers of bins

Add new features columns to the train and test sets

```python
for feautre, bins in zip(features_to_transform, optimal_bins):
    train_set=numeric_to_categorical(train_set, feautre, bins=bins)
    test_set=numeric_to_categorical(test_set, feautre, bins=bins)

train_set.columns
```

## 2.6 Save new datasets

```python
train_set.to_csv("./preprocessed_data/train_set_stage2.csv")
test_set.to_csv("./preprocessed_data/test_set_stage2.csv")
```
